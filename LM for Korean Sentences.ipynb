한국어에서의 언어 모델(LM for Korean Sentences)
===
* 한국어는 영어나 기타 언어에 비해서 언어 모델로 다음 단어를 예측하기 훨씬 까다롭다.

1. 한국어는 어순이 중요하지 않다.
---
* 한국어에서는 어순이 중요하지 않다. 이전 단어가 주어졌을 때, 다음 다어가 나타날 확률을 구해야하는데 어순이 중요하지 않다는 것은 어떤 단어든 나타나도 된다는 의미
* 예를들어 순서를 뒤죽박죽으로 바꾸어놔도 한국어는 의미가 전달되기 때문에 확률에 기반한 언어 모델이 제대로 다움 단어를 예측하기가 어렵습니다.

2. 한국어는 교착어이다.
---
* 띄어쓰기 단위인 어절 단위로 토큰화를 할 경우에는 문장에서 발생가능한 단어의 수가 굉장히 늘어난다.
* 예를들어 교착어인 한국어에는 조사가 있다. 영어는 기본적으로 조사가 없지만 한국어에는 어떤 행동을 하는 동사의 주어나 목적어를 위해서 조사라는 것이 있다.

3. 한국어는 띄어쓰기가 제대로 지켜지지 않는다.
---
* 한국어는 띄어쓰기를 제대로 하지않아도 의미가 전달되고, 규칙 또한 상대적으로 까다롭다.
* 자연어 처리에서 한국어 코퍼스는 띄어쓰기가 제대로 지켜지지 않는 경우가 많다. 그래서 토큰이 제대로 분리되지 않은 채 훈련 데이터로 사용된다면, 언어 모델은 제대로 동작하지 않는다.
